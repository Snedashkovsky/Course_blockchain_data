### Get config

```python
rpc.get_config()
```

### Get dynamic global properties

```python
rpc.get_dynamic_global_properties()
```

```python
last_block = rpc.get_dynamic_global_properties()['current_aslot']
print('Last block: {}'.format(last_block))
```
